```python
df_ash["price"] = df_ash["price"].replace('[\$,]', '', regex=True).astype(float)
```

```python
df_ash = df_ash.dropna(subset=["price"])
```

```python
print(df_ash["price"].isnull().sum())
```

```python
df_ash['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df_ash['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())

print(df_ash["bedrooms"].isnull().sum())
print(df_ash["bathrooms"].isnull().sum())
```

```python
df_ash["review_scores_rating"].isnull().sum()
```

```python
df_ash["review_scores_rating"] = df_ash["review_scores_rating"].fillna(0)
df_ash["has_review_scores"] = (df_ash["review_scores_rating"] > 0).astype(int)
```

```python
df_ash["review_scores_rating"].isnull().sum()
```

```python
df_ash["has_review_scores"].value_counts()
```

```python
df_model=df_ash[["accommodates","bedrooms","bathrooms","review_scores_rating","minimum_nights","has_review_scores"]]
```

```python
df_model.reset_index(drop=True, inplace=True)  # Reset index first
df_model.index = df_model.index + 1 
df_model
```

```python
correlation_matrix=df_model.corr()
correlation_matrix
```

```python
import seaborn as sns 
plt.figure(figsize=(15, 6))
heatmap=sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 0, fontsize = 10)
plt.title("Correlation Matrix")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x=df_model
y=df_ash["price"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R2): {r2:.2f}')
```
