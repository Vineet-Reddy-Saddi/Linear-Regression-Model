Removing the $symbol from the price and coverting the data type to float to perform furthur calculations

```python
df_ash["price"] = df_ash["price"].replace('[\$,]', '', regex=True).astype(float)
```

```python
df_ash = df_ash.dropna(subset=["price"])
```

As both the bedrooms and bathrooms columns are skewed we used median to fill the missing values

```python
df_ash['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df_ash['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())

print(df_ash["bedrooms"].isnull().sum())
print(df_ash["bathrooms"].isnull().sum())
```

```python
df_ash["review_scores_rating"].isnull().sum()
```

Creating dummy variables for review_score_Rating and creating a new column has_Review_Scores with values 0 if there is no rating and 1 if there is some rating 

```python
df_ash["review_scores_rating"] = df_ash["review_scores_rating"].fillna(0)
df_ash["has_review_scores"] = df_ash["review_scores_rating"].apply(lambda x: 1 if x > 0 else 0)
```

# 3. Model Building

Importing the neccessary machine learning libraries to build the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import StandardScaler
```

A.	Spliting the data into training (80%) and testing (20%) sets to build a linear regression model using five features and the dummy variable (has_review_scores).

```python
x=df_model[["accommodates","bedrooms","bathrooms","review_scores_rating","minimum_nights","has_review_scores"]]
y=df_model["price"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)


model=LinearRegression()
model.fit(x_train_scaled,y_train)
```

B. Predicting the values and evaluating the mean squared error and R squared error 

```python
y_pred=model.predict(x_test_scaled)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R2): {r2:.2f}')
```
