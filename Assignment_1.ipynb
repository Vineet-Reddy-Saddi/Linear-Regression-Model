```python
df_ash["price"] = df_ash["price"].replace('[\$,]', '', regex=True).astype(float)
```

```python
df_ash.info()
```

```python
df_ash = df_ash.dropna(subset=["price"])
```

```python
print(df_ash["price"].isnull().sum())
```

```python
df_ash['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df_ash['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())

print(df_ash["bedrooms"].isnull().sum())
print(df_ash["bathrooms"].isnull().sum())
```

```python
df_ash["review_scores_rating"].isnull().sum()
```

```python
df_ash["review_scores_rating"] = df_ash["review_scores_rating"].fillna(0)
df_ash["has_review_scores"] = (df_ash["review_scores_rating"] > 0).astype(int)
```
